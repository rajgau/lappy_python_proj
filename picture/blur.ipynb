```python

maper = lambda var,minI,maxI,minO,maxO:(((var-minI)/(maxI-minI))*(maxO-minO))+minO
dist = lambda x,y,x1,y1:((x1-x)**2+(y1-y)**2)**.5
plt.imshow(img)
```

```python
grid=1
max_dist=dist(grid,grid,0,0)
a=np.array([[maper(dist(i,j,grid,grid),0,max_dist,1,0) for j in range(2*grid+1)] for i in range(2*grid+1)])
plt.imshow(a)
```

```python
img2=np.zeros(img.shape)
width=len(img)
height=len(img[0])
for i in range(width):
    for j in range(height):
        for k in range(3):
            sumer=0
            for l in range(-grid,grid+1):
                if 0<i+l<width:
                    for m in range(-grid,grid+1):
                        if 0<j+m<height:
                            sumer+=img[i+l][j+m][k]
            c=sumer/((2*grid+1)**3)
            img2[i][j][k]=c
```
