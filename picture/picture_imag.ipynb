```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
from PIL import Image as im
from tensorflow import keras
```

```python
def blur(img,col=False):
    img1=[]
    if not col:
        for i in range(28):
            ad=[]
            for j in range(28):
                sumer=0
                for x in range(-1,2):
                    for y in range(-1,2):
                        if i+x<0 or i+x>27 or j+y<0 or j+y>27:continue
                        sumer+=img[i+x][j+y]/9
                ad.append(sumer)
            img1.append(ad)
    else:
        for i in range(len(img)):
            ad=[]
            for j in range(len(img[0])):
                ck=[]
                for k in range(3):
                    sumer=0
                    for x in range(-1,2):
                        for y in range(-1,2):
                            if i+x<0 or i+x>len(img)-1 or j+y<0 or j+y>len(img[0])-1:continue
                            sumer+=img[i+x][j+y][k]/(255*9)
                    ck.append(sumer)
                ad.append(ck)
            img1.append(ad)
    return img1
def rec_blur(img,n,col=False):
    for i in range(n):img=blur(img,col)
    return img
```
