```python
from PIL import Image as im
import matplotlib.pyplot as plt
```

```python
img = im.open("0.jpg").resize((500,300)).__array__()
```

```python
def blur(img,fac):
    side=(2*fac)+1
    off = [[1 for j in range(side)] for i in range(side)]
    for i in range(len(img[0])):
        for j in range(len(img)):
            for k in range(3):
                try:
                    sumer=0
                    for x,l in enumerate(off):
                        for y,m in enumerate(l):
                            if j+(x-fac)<0 or j+(x-fac)==len(img):sumer+=0
                            elif i+(y-fac)<0 or i+(y-fac)==len(img):sumer+=0
                            else:sumer+=img[j+(x-fac),i+(y-fac),k]*(m/(side**2))
                    img[j,i,k]=int(sumer)
                except:img[j,i,k]=50*(k-2)
    return img
```
